```python
import s3fs
import os
```

```python
src_test_img_path = 'dev-pdi-training-images/footprint/from-elu-sm-instance/test-images'
dest_input_img_path = 'dev-pdi-training-images/footprint-new-structure/dev-test-images/input-images'
dest_gt_img_path = 'dev-pdi-training-images/footprint-new-structure/dev-test-images/gt-images'
```

```python
fs = s3fs.S3FileSystem()
```

```python
lst_test_imgs = fs.ls(dest_input_img_path)
len(lst_test_imgs)
```

```python
lst_test_imgs = fs.ls(dest_gt_img_path)
len(lst_test_imgs)
```

```python
img = os.path.split(lst_test_imgs[0])
```

```python
img
```

```python
img[0].find('jpg')
```

```python
input_knt = 1
gt_knt = 1
for idx, img in enumerate(lst_test_imgs, start=1):
    src_img = os.path.split(img)[1]
    print('%d) %s' % (idx, src_img))
    #check whether it is label or input img.
    if(src_img.find('_bfp') > 0):
        #Label Image
        dest_lbl_path = os.path.join(dest_gt_img_path, src_img)
        fs.copy(img, dest_lbl_path)
        gt_knt += 1
    else:
        #Input Image
        dest_input_path = os.path.join(dest_input_img_path, src_img)
        fs.copy(img, dest_input_path)
        input_knt += 1

```

```python
print('Total Input Images : {} '.format(input_knt))
print('Total Label Images : {} '.format(gt_knt))
```
