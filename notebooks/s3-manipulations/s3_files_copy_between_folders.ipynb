```python
import shutil
import s3fs
import os
```

```python
fs = s3fs.S3FileSystem()
```

```python
#image_1.tif
input_img_path = 'dev-pdi-training-images/footprint-new-structure/training-images/input-images/base-9831-input-images/'
label_src_path = 'dev-pdi-training-images/footprint/Version_1.0/label/'   #image_10000_gt.tif
label_dest_path = 'dev-pdi-training-images/footprint-new-structure/training-images/gt-images/base-9831-gt-images/'
```

```python
lst_input_imgs = fs.ls(input_img_path)
len(lst_input_imgs), lst_input_imgs[0]
```

```python
os.path.split(lst_input_imgs[0])[1]
```

```python
for idx, input_img in  enumerate(lst_input_imgs, start=1):
    input_file_name, _ = os.path.splitext(os.path.split(input_img)[1])
    lbl_filename = input_file_name + '_gt.tif'
    input_file_name += '.tif'
    print("%d ) %s, \t %s" % (idx, input_file_name, lbl_filename))
    
    #copy files 
    src_label_file = os.path.join(label_src_path, lbl_filename)
    dest_label_file = os.path.join(label_dest_path, lbl_filename)
    fs.copy(src_label_file, dest_label_file)
    
```
