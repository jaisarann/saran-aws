```python
import cv2 as cv
import os
from matplotlib import pyplot as plt
```

```python
gt_folder_path = r'/home/ec2-user/SageMaker/data/footprint/dorian-finalized-178-images/gt'
gt_target_path = r'/home/ec2-user/SageMaker/data/footprint/dorian-finalized-178-images/gt-binary'
```

```python
gt_labels = [os.path.join(gt_folder_path, fl) for fl in os.listdir(gt_folder_path)]
gt_labels
```

```python
for idx, lbl_img in enumerate(gt_labels, start=1):
#    print(lbl_img)
    img = cv.imread(lbl_img)
    gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    gray_img[gray_img > 0] = 1
    _, lbl_file_name_with_extn = os.path.split(lbl_img)
    cv.imwrite(os.path.join(gt_target_path, lbl_file_name_with_extn), gray_img)
    print('processing', idx)
    
print('done')
```
